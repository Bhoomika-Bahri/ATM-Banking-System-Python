####Imports and File Setup

```python
import json
import os

FILE_NAME = "bank_data.txt"
```

#### Load and Save Functions

```python
def load_data():
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r") as file:
            return json.load(file)
    return {}

def save_data(data):
    with open(FILE_NAME, "w") as file:
        json.dump(data, file)
```

#### Account Creation

```python
def create_account(data):
    name = input("Enter your name: ")
    pin = input("Set a 4-digit PIN: ")

    acc_no = str(len(data) + 1001)

    data[acc_no] = {
        "name": name,
        "pin": pin,
        "currency": "EUR",
        "balance": 0,
        "transactions": []
    }

    print(f"\nAccount created successfully!")
    print(f"Your Account Number is: {acc_no}")
    return data
```

#### Login Function

```python
def login(data):
    acc_no = input("Enter account number: ")
    pin = input("Enter PIN: ")

    if acc_no in data and data[acc_no]["pin"] == pin:
        print("\nLogin successful!\n")
        return acc_no
    else:
        print("Invalid account number or PIN.")
        return None
```

####Deposit and Withdraw Function

```python
def deposit(data, acc_no):
    amount = float(input("Enter amount to deposit (€): "))
    data[acc_no]["balance"] += amount
    data[acc_no]["transactions"].append(f"Deposited €{amount}")
    print("Deposit successful!")


def withdraw(data, acc_no):
    amount = float(input("Enter amount to withdraw (€): "))

    if amount <= data[acc_no]["balance"]:
        data[acc_no]["balance"] -= amount
        data[acc_no]["transactions"].append(f"Withdrawn €{amount}")
        print("Withdrawal successful!")
    else:
        print("Insufficient balance!")
```

#### Balance and Account Details

```python
def check_balance(data, acc_no):
    print(f"Current Balance: €{data[acc_no]['balance']}")


def account_details(data, acc_no):
    acc = data[acc_no]

    print("\n--- ACCOUNT DETAILS ---")
    print("Name:", acc["name"])
    print("Account Number:", acc_no)
    print("Currency: EUR (€)")
    print("Balance: €", acc["balance"])

    print("\nTransactions:")
    for t in acc["transactions"]:
        print("-", t)
```

#### **User** **Menu**

```python
def user_menu(data, acc_no):
    while True:
        print("\n===== ATM MENU =====")
        print("1. Deposit")
        print("2. Withdraw")
        print("3. Check Balance")
        print("4. Account Details")
        print("5. Logout")

        choice = input("Enter choice: ")

        if choice == "1":
            deposit(data, acc_no)

        elif choice == "2":
            withdraw(data, acc_no)

        elif choice == "3":
            check_balance(data, acc_no)

        elif choice == "4":
            account_details(data, acc_no)

        elif choice == "5":
            print("Logging out...\n")
            break

        else:
            print("Invalid choice")

        save_data(data)
```

#### **Application Runner**

```python
data = load_data()

while True:
    print("\n===== WELCOME TO ATM SYSTEM =====")
    print("1. Create Account")
    print("2. Login")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        data = create_account(data)
        save_data(data)

    elif choice == "2":
        acc_no = login(data)

        if acc_no:
            user_menu(data, acc_no)

    elif choice == "3":
        save_data(data)
        print("Thank you for using ATM system!")
        break

    else:
        print("Invalid choice")
```
